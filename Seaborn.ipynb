```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips=sns.load_dataset('tips') #tips set is already available
tips
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) 
#by default aggregate function is mean but with  the help of estimator we can select aggregate function for 
#example-std deviation
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.distplot(tips['total_bill'],bins=40)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='day',y='total_bill',data=tips) # to visualize two variables or graph
```

```python
sns.pairplot(tips)
```

```python
tips_correlation=tips.corr()
print(tips_correlation)
```

```python
sns.heatmap(tips.corr(),annot=True)
```
